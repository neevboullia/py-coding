## 期中專題

#### 我常常在玩猜數字遊戲 1A2B，大部分都在第五次或第六次猜出答案，
#### 因此我在想，這樣算不算猜得快，以及有什麼猜測策略能降低猜出答案的期望值，因此有此題目

#### 以下分為兩個部分，第一個部分為自製的猜數字遊戲供大家試玩
#### 第二部分為使用最簡單的策略測試猜測次數的期望值

  

#### 一、建立遊戲 猜數字 1A2B

規則 猜四位不重復數字數字

猜對且位置對為A

猜對但位置錯為B

例如

答案5712 

猜測1732 為2A1B

```python
import pandas as pd
import numpy as np
from itertools import combinations, permutations #排列組合
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
answer = list(np.random.permutation(10)[0:4])  # 首先必須產生一組數字不重複的四位數隨機數字 (此為答案)
count = 1                                      # 方法 : 將0-9隨機排列取前四位數
t1= time.time()    
while True:    
    print("第",count,"次猜測")
    g = input('Please input 4 different numbers :')
    gu = str(g)
    if len(gu) != 4:     #輸入不合4位數
        print("please input 4 numbers")
        break
    gue = [ gu[i] for i in range(0,4)]
    guess = [ int(gue[i]) for i in range(0,4)]
    if len(np.unique(guess)) != 4:  #輸入位數重複
        print("please input 4 different numbers like 7821")
        break
    A = sum([ guess[i] == answer[i] for i in range(0,4)])  #猜對且位置對
    B = sum([ guess[i] in answer for i in range(0,4)])-A   #猜對但位置錯
    if A == 4:
        print("恭喜答對!!")
        break
    else:
        print(f"{A} A {B} B ")
    count+=1
    t2 = time.time()
    print(f'總花費的時間：{np.round(t2 - t1)}s') #計算時間
```

#### 二、檢驗猜測次數期望值

##### 這裡使用最簡單的策略，猜錯後隨機從剩下的可能性中挑選下次的猜測


首先列出所有可能性，利用itertools套件做排列組合

```python
database = list(permutations([1,2,3,4,5,6,7,8,9,0],4))   #5040種 (P 10取4)  
#list 包含5040個tuple
```

接著產生一組數字不重複的四位數隨機數字 (此為答案)

由於database順序固定 ，為了確保隨機，因此答案不從database取。

方法 : 將0-9隨機排列取前四位數

```python
n= []
for q in range(0,1000):

    database = list(permutations([1,2,3,4,5,6,7,8,9,0],4))
    answer = list(np.random.permutation(10)[0:4])
    guess = list(np.random.permutation(10)[0:4])   # 接著隨機猜一組數字
    k = 1 #th guess

    while True: 

        A = sum([ guess[i] == answer[i] for i in range(0,4)])  #猜對且位置對
        B = sum([ guess[i] in answer for i in range(0,4)])-A   #猜對但位置錯

        print(f"第{k}次猜測:{guess},結果{A} A {B} B")
        if A == 4:
            break
        checkA = []     #接著從資料庫篩選符合的組合
        checkb = []
        for j in range(0,len(database)):
            checkA.append(sum([ guess[i] == database[j][i] for i in range(0,4)]))   #猜對且位置對
            checkb.append(sum([ guess[i] in database[j] for i in range(0,4)]))      #猜對但位置錯
        checkB = list(np.array(checkb)-np.array(checkA))

        check = [(checkA[i]==A) & (checkB[i]==B)  for i in range(0,len(checkA))]  #將符合標準取出

        newdatabase = list(np.array(database)[np.array(check)])  #np.array 做邏輯運算
        newdatabase = [ list(newdatabase[i])  for i in range(0,sum(check))]  #換回List
        guess = newdatabase[np.random.randint(0,sum(check))]   #下次猜測
        database = newdatabase
        k+=1
    n.append(k)
```

```python
tt = np.array(n)
ac = [sum(tt==i) for i in range (3,9)]
m = np.mean(tt)
plt.hist(tt)
plt.xlabel('guess times')
```

```python
print(f"平均猜測{m}次")
print(f"在第5、6次猜出的機率為{(ac[2]+ac[3])/1000}")
```

### 結論

猜測次數大致上呈現常態分布，

如以此策略猜則期望值落在5.5次左右(1000次試驗)，

而在第5、6次猜出的機率為0.713

因此若沒在5次以內猜出來則可能是策略不好，

又或者是這個遊戲其實沒有好的策略。
