```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact;
```

```python
x = np.linspace(0,6,1000)
y = (x-1)*(x-2)*(x-3)*(x-5) 
# y = x^4 - 11x^3 + 41x^2 - 61x + 30
# y' = 4x^3 -33x^2 + 82x -61
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-10,10)
plt.title('loss function')
```

#### 設計一函數能決定起始點origin，以及改變的速率learning_rate

```python
def loss_fun(origin = 3,learning_rate = 0.1):   
    k = 0
    t = origin  #起始點
    x = np.linspace(0,6,1000)
    y = (x-1)*(x-2)*(x-3)*(x-5) 
    plt.plot(x,y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.ylim(-10,10)
    plt.title('loss function')
    while True:
        k+=1
        move = learning_rate / k  #移動逐漸減少
        slope = 4*t**3 -33*t**2 + 82*t -61   #一次微分 : 斜率
        y = (t-1)*(t-2)*(t-3)*(t-5) 

        plt.scatter(t,y)
        t -= slope*move   #以斜率大小當作移動的多寡
        
        if abs(slope) < 0.01:   #當斜率趨近於零時停止
            print("end ")
            y = (t-1)*(t-2)*(t-3)*(t-5) 
            plt.scatter(t,y)
            break
```

```python
interact(loss_fun, origin=(0, 6.),learning_rate = (0.01,1.))
```

結果發現，當learning_rate設定的不好會有異常狀況，

learning_rate太小 : 無法達到最小值

learning_rate太大 : 變動的值會越過local maximum，而導致找到另一個local minimum
